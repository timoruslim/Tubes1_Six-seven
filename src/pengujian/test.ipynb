```python
%reload_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

from ffnn.nn import MLP, Layer, RMSNorm
```

```python
X, y = make_moons(n_samples=300, noise=0.15, random_state=42)
y = y.reshape(-1, 1).astype(float) 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=67)
print(X_train.shape, y_train.shape)
```

```python
model = MLP([
    Layer(16, activation='relu', weight_init='he'), 
    Layer(8, activation='relu', weight_init='he'),
    RMSNorm(8),
], input_size=2, seed=67)
```

```python
model.add(Layer(1, activation='sigmoid', weight_init='xavier'))
```

```python
model.summary()
```

```python
model.compile(optimizer={'method': 'adam', 'learning_rate': 0.008}, loss='bce')

history = model.fit(
    X_train, y_train, 
    batch_size=16, 
    epochs=6000, 
    validation_data=(X_test, y_test), 
    verbose=1
)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['val_loss'], label='Validation Loss')
plt.title('Learning Curve (Adam)')
plt.xlabel('Epoch')
plt.ylabel('BCE Loss')
plt.legend()
plt.show()
```

```python
model.plot_weights(layer_indices=[0, 1, 2]) 
model.plot_gradients(layer_indices=[0, 1, 2])
```
